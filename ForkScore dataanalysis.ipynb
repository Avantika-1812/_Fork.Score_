Roberta Pretrained Model

- Use a model trained of a large corpus of data.
- Transformer model accounts for the words but also the context related to other words.

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
# VADER results on example
print(example)
sia.polarity_scores(example)
```

```python
# Run for Roberta Model
encoded_text = tokenizer(example, return_tensors='pt')
output = model(**encoded_text)
scores = output[0][0].detach().numpy()
scores = softmax(scores)
scores_dict = {
    'roberta_neg' : scores[0],
    'roberta_neu' : scores[1],
    'roberta_pos' : scores[2]
}
print(scores_dict)
```

```python
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['Text']
        myid = row['Id']
        vader_result = sia.polarity_scores(text)
        vader_result_rename = {}
        for key, value in vader_result.items():
            vader_result_rename[f"vader_{key}"] = value
        roberta_result = polarity_scores_roberta(text)
        both = {**vader_result_rename, **roberta_result}
        res[myid] = both
    except RuntimeError:
        print(f'Broke for id {myid}')
```

```python
results_df = pd.DataFrame(res).T
results_df = results_df.reset_index().rename(columns={'index': 'Id'})
results_df = results_df.merge(df, how='left')
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming your results are stored in 'results_df' as before

# 1. VADER Performance
# a. Correlation
vader_correlation = results_df[['vader_compound', 'Score']].corr().iloc[0, 1]
print(f"VADER Correlation with Score: {vader_correlation}")

# b. Categorization and Accuracy Metrics
results_df['vader_sentiment'] = pd.cut(results_df['vader_compound'],
                                      bins=[-1, -0.05, 0.05, 1],
                                      labels=['negative', 'neutral', 'positive'])
results_df['ground_truth_sentiment'] = pd.cut(results_df['Score'],
                                             bins=[0, 2, 4, 5],
                                             labels=['negative', 'neutral', 'positive'])

# Calculate accuracy and other metrics
vader_accuracy = accuracy_score(results_df['ground_truth_sentiment'], results_df['vader_sentiment'])
vader_precision = precision_score(results_df['ground_truth_sentiment'], results_df['vader_sentiment'], average='weighted')
vader_recall = recall_score(results_df['ground_truth_sentiment'], results_df['vader_sentiment'], average='weighted')
vader_f1 = f1_score(results_df['ground_truth_sentiment'], results_df['vader_sentiment'], average='weighted')

print(f"VADER Accuracy: {vader_accuracy}")
print(f"VADER Precision: {vader_precision}")
print(f"VADER Recall: {vader_recall}")
print(f"VADER F1-Score: {vader_f1}")

# 2. Roberta Performance (Similar approach)
# a. Correlation
roberta_correlation = results_df[['roberta_pos', 'Score']].corr().iloc[0, 1]
print(f"Roberta Correlation with Score (using roberta_pos): {roberta_correlation}")

# b. Categorization and Accuracy Metrics
results_df['roberta_sentiment'] = results_df[['roberta_neg', 'roberta_neu', 'roberta_pos']].idxmax(axis=1)
results_df['roberta_sentiment'] = results_df['roberta_sentiment'].map({'roberta_neg': 'negative', 'roberta_neu': 'neutral', 'roberta_pos': 'positive'})

# Calculate accuracy and other metrics
roberta_accuracy = accuracy_score(results_df['ground_truth_sentiment'], results_df['roberta_sentiment'])
roberta_precision = precision_score(results_df['ground_truth_sentiment'], results_df['roberta_sentiment'], average='weighted')
roberta_recall = recall_score(results_df['ground_truth_sentiment'], results_df['roberta_sentiment'], average='weighted')
roberta_f1 = f1_score(results_df['ground_truth_sentiment'], results_df['roberta_sentiment'], average='weighted')

print(f"Roberta Accuracy: {roberta_accuracy}")
print(f"Roberta Precision: {roberta_precision}")
print(f"Roberta Recall: {roberta_recall}")
print(f"Roberta F1-Score: {roberta_f1}")
```
